```python
search_input_element = driver.find_element_by_name("keyword")
search_input_element.send_keys(keyword_1)
search_button_element = driver.find_element_by_tag_name("button")
search_button_element.click()
```

```python
pageSource = driver.page_source
bsbj = BeautifulSoup(pageSource, 'html.parser')
```

```python
all = []
##page 1 - 10 개 까지 총 200개
for i in range (1,11):
    ppompage = "http://www.ppomppu.co.kr/search_bbs.php?search_type=sub_memo&page_no="+str(i)+"&keyword=%B8%F4%C5%D7%C0%CF&page_size=20&bbs_id=&order_type=date&bbs_cate=1"## 중간에 str 입력시 ' ,  "  앞에 쓴거 동일하게 해야함
    html = urlopen(ppompage)
    Beautiful = BeautifulSoup(html,"html.parser")
    basic_infor = Beautiful.find_all('p',{"class":"desc"})
    seper = []
    for b in basic_infor:
        bbb = b.get_text().strip().split(sep='\n')
        seper.append(bbb)
    seper_2 = []
    seper_2_2 = []
    for c in Beautiful.select(".content > .title > a"):
        seper_2.append(c.text)
        seper_2_2.append(c["href"])
    for d in range(1,21):
        seper[d-1].append(seper_2[d-1])
        seper[d-1].append(seper_2_2[d-1])
    all.extend(seper[0:20])

```

```python
text1 = []
for i in range(1,201):
    path_1 = all[i-1][6]
    driver.get(path_1)
    source1 = driver.page_source
    Bsoup1 = BeautifulSoup(source1,"html.parser")
    content1 = Bsoup1.find_all('td', {"class" : "board-contents"})
    for r in content1:
        m = r.get_text().strip().replace("\n","").split(sep = "\n")
        text1.append(m)
all_2 = all
for x in range(1,201):
    all_2[x-1].append(text1[x-1])
```
