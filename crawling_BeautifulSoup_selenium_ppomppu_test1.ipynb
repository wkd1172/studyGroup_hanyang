```python
all = []

for i in range (1,3):
    ppompage = "http://www.ppomppu.co.kr/search_bbs.php?search_type=sub_memo&page_no="+str(i)+"&keyword=%B8%F4%C5%D7%C0%CF&page_size=20&bbs_id=&order_type=date&bbs_cate=1"## 중간에 str 입력시 ' ,  "  앞에 쓴거 동일하게 해야함
    html = urlopen(ppompage)
    ## 보기 계층별 분류
    Beautiful = BeautifulSoup(html,"html.parser")
    ## 조회수, 날짜등 불러오기
    basic_infor = Beautiful.find_all('p',{"class":"desc"})
    ## basic_infor 를 정보에 따라 분류하기
        #하나의 basic_infor의 객체를 
    seper = []
    for b in basic_infor:
        bbb = b.get_text().strip().split(sep='\n')  ### strip 는 문자열의 양 쪽 공백을 지우는것
                                            ### split("")  안에 있는 것으로 나누는것
        seper.append(bbb) ### append 함수는 변수에 차래대로 값을 추가하는 것이다.
    ## 제목 링크 가져오기
    seper_2 = []
    seper_2_2 = []
    for c in Beautiful.select(".content > .title > a"): #### select로 나온 결과는 제목만 나온다.
        seper_2.append(c.text) #### 변수.text는 제목, 내용같은 것만 글어온다
        seper_2_2.append(c["href"]) ### c중에서 href  = 에 해당되는 것만 글어온다.
    ## 합치기
    for d in range(1,21):#### range (n,m) = range n부터 m-1 까지다
        seper[d-1].append(seper_2[d-1])
        seper[d-1].append(seper_2_2[d-1])
    all.extend(seper[0:20]) ### for 문 밖에 golobal 변수가 필요하다. 거기다가. 각 각 집어 넣기 ! 
                            ### append 와 extend 차이점!!
#################### all 에 합쳐져 있다.
    ## all[6] 에 링크가 되어있다.!!

##### oversee 해외포럼 과 다른 2가지 글은 text가 있는 곳이 다른가 보다.
        
            
all
```

```python
### 왜 이렇게 하면 해외포럼 결과 밖에 아난오는지 이제 이해됨. 
### 나머지 해외장터 같은 곳은 로그인이 필요한 곳이기떄문에 driver를 무조건 이용해야 됨.

for i in range(1,41):
    page_source = all[i-1][6]
    if 'oversea' in page_source:
        urlopen_2 = urlopen(page_source)
        Bsoup_2 = BeautifulSoup(urlopen_2,"html.parser")
        content_2 = Bsoup_2.find_all(class_ = 'pic_bg')
        for w in content_2:
            print(w.get_text().strip().replace("\n","").split(sep = "\n"))
    elif 'market'in page_source:
        urlopen_3 = urlopen(page_source)
        Bsoup_3 = BeautifulSoup(urlopen_3,"html.parser")
        content_3 = Bsoup_3.find_all('td', {"class" : "board-contents"})
        for r in content_3:
            print(r.get_text().strip().replace("\n",""),split(sep = "\n"))
    elif 'relay' in page_source:
        urlopen_4 = urlopen(page_source)
        Bsoup_4 = BeautifulSoup(urlopen_4,"html.parser")
        content_4 = Bsoup_4.find_all('td', {"class" : "board-contents"})
        for m in content_4:
            print(m.get_text().strip().replace("\n",""),split(sep = "\n"))
```
